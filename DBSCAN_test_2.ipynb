# Import Dependencies

```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.decomposition import PCA
from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
%matplotlib inline
```

# Import Data

```python
df = pd.read_csv('../Data/action_log_features.csv')
```

```python
df.head()
```

```python
users = df[['account', 'at']].apply(tuple, axis=1).tolist()
df['account'] = users
df.drop('at', inplace=True, axis=1)
df.rename(columns={'account':'user'}, inplace=True)

df.set_index('user', inplace=True)
```

```python
df.head()
```

# Standardizing data

```python
x = df.values # numpy array
scaler = preprocessing.MinMaxScaler()
x_scaled = scaler.fit_transform(x)
X_norm = pd.DataFrame(x_scaled)
```

```python
X_norm
```

# Use PCA to reduce 60 columns into 2

```python
pca = PCA(n_components = 2) # 2D PCA for the plot
reduced = pd.DataFrame(pca.fit_transform(X_norm))
```

# Apply DBSCAN

```python
# train the model using DBSCAN
db = DBSCAN(eps=1, min_samples=5)

# prediction for dbscan clusters
db_clusters = db.fit_predict(reduced)
```
