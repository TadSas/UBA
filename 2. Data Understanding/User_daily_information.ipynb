# GroupBy and Extract Date, Aim Tuple

```python
def group_and_extract(df, sdf, account=2):
    grp = df[df['account'] == account].groupby('at')
    
    inner_data = grp['aim'].value_counts().index.get_level_values('at')
    data = sorted(list(set(map(lambda x: str(x)[:10], inner_data))))
    
    for date in data:
        tp = grp['aim'].value_counts()[date]
        lst = list(zip(tp.index, tp))
        add_data_to_df(sdf, date, lst, account)
    sdf.fillna(0, inplace=True)
    return sdf
```

# Add Data To SDF

```python
def add_data_to_df(sdf, date, lst, account):
    
    cols = [el[0] for el in lst]
    vals = [el[1] for el in lst]
    cols.insert(0, 'date')
    cols.insert(1, 'user_id')
    vals.insert(0, date)
    vals.insert(1, account)
    if sdf.empty:
        sdf.loc[0, cols] = vals
    else:
        sdf.loc[sdf.index[-1] + 1, cols] = vals
```
